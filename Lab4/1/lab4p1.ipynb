```python
import pandas as pd
import numpy as np
from datetime import datetime
import time
```

### Завантаження данних

```python
file_path = "household_power_consumption.txt"
```

Pandas

```python
def load_data_pandas(file_path):
    data = pd.read_csv(file_path, sep=';', na_values=['?'], low_memory=False)
    data.dropna(inplace=True)
    # data['DateTime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'], format='%d/%m/%Y %H:%M:%S')
    return data
```

```python
data_p = load_data_pandas(file_path)
data_p
```

NumPy

```python
def load_data_numpy(file_path):

    types = [
        ("Date", "U10"),
        ("Time", "U8"),
        ("Global_active_power", "float64"),
        ("Global_reactive_power", "float64"),
        ("Voltage", "float64"),
        ("Global_intensity", "float64"),
        ("Sub_metering_1", "float64"),
        ("Sub_metering_2", "float64"),
        ("Sub_metering_3", "float64")
    ]
    
    data = np.genfromtxt(
        file_path, 
        delimiter=';', 
        dtype=types, 
        missing_values=["?", "nan"], 
        encoding="UTF-8", 
        names=True
    )
    
    return data
```

```python
data_n = load_data_numpy(file_path)
data_n
```

### Завдання 3

Обрати всі записи, у яких сила струму лежить в межах 19-20 А, для них виявити ті, у яких пральна машина та холодильних споживають більше, ніж бойлер та кондиціонер.

Pandas

```python
def task3_pandas(data):
    subset = data[(data['Global_intensity'] >= 19) & (data['Global_intensity'] <= 20)]
    return subset[subset['Sub_metering_2'] > subset['Sub_metering_3']]
```

```python
data_p3 = task3_pandas(data_p)
data_p3
```

NumPy

```python
def task3_numpy(data):
    intensity = data['Global_intensity'].astype(float)
    mask = (intensity >= 19) & (intensity <= 20)
    subset = data[mask]
    sub2 = subset['Sub_metering_2'].astype(float)
    sub3 = subset['Sub_metering_3'].astype(float)
    return subset[sub2 > sub3]
```

```python
data_n3 = task3_numpy(data_n)
data_n3
```

### Завдання 4

Обрати випадковим чином 500000 записів (без повторів елементів вибірки), для них обчислити середні величини усіх 3-х груп споживання електричної енергії

Pandas

```python
def task4_pandas(data):
    sample = data.sample(500000, random_state=42)
    means = sample[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean()
    return means
```

```python
data_p4 = task4_pandas(data_p)
data_p4
```

NumPy

```python
def task4_numpy(data):
    sample_indices = np.random.choice(data.shape[0], 500000, replace=False)
    sample = data[sample_indices]
    
    sub1 = sample['Sub_metering_1']
    sub2 = sample['Sub_metering_2']
    sub3 = sample['Sub_metering_3']
    
    sub1_mean = sub1[~np.isnan(sub1)].mean()
    sub2_mean = sub2[~np.isnan(sub2)].mean()
    sub3_mean = sub3[~np.isnan(sub3)].mean()
    
    return sub1_mean, sub2_mean, sub3_mean
```

```python
data_n4 = task4_numpy(data_n)
data_n4
```

### Завдання 5

Обрати ті записи, які після 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких основне споживання електроенергії у вказаний проміжок часу припадає на пральну машину, сушарку, холодильник та освітлення (група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другої половини.

Pandas

```python
def task5_pandas(data):
    data['Hour'] = data['Time'].str.split(':').str[0].astype(int)
    after_6pm = data[(data['Hour'] >= 18) & (data['Global_active_power'] > 6)]
    group2_max = after_6pm['Sub_metering_2'] > after_6pm[['Sub_metering_1', 'Sub_metering_3']].max(axis=1)
    filtered = after_6pm[group2_max]
    first_half = filtered.iloc[:len(filtered)//2]
    second_half = filtered.iloc[len(filtered)//2:]
    return pd.concat([first_half.iloc[::3], second_half.iloc[::4]])
```

```python
data_p5 = task5_pandas(data_p)
data_p5
```

NumPy

```python
def task5_numpy(data):
    times = np.array([datetime.strptime(d + ' ' + t, "%d/%m/%Y %H:%M:%S") for d, t in zip(data['Date'], data['Time'])])
    hours = np.array([t.hour for t in times])
    mask = (hours >= 18) & (data['Global_active_power'] > 6)
    filtered = data[mask]
    
    group2_dominance = filtered['Sub_metering_2'] > np.maximum(filtered['Sub_metering_1'], filtered['Sub_metering_3'])
    filtered = filtered[group2_dominance]
    
    midpoint = len(filtered) // 2
    first_half = filtered[:midpoint]
    second_half = filtered[midpoint:]
    
    first_half_sample = first_half[::3]
    second_half_sample = second_half[::4]
    
    result = np.concatenate((first_half_sample, second_half_sample))
    return result
```

```python
data_n5 = task5_numpy(data_n)
data_n5
```

### Порівняння швидкості

```python
results = {}

for i, (pandas_func, numpy_func, name) in enumerate([
    (task1_pandas, task1_numpy, "Task 1"),
    (task2_pandas, task2_numpy, "Task 2"),
    (task3_pandas, task3_numpy, "Task 3"),
    (task4_pandas, task4_numpy, "Task 4"),
    (task5_pandas, task5_numpy, "Task 5")
]):
    
    start = time.time()
    pandas_result = pandas_func(data_p)
    pandas_time = time.time() - start

    start = time.time()
    numpy_result = numpy_func(data_n)
    numpy_time = time.time() - start

    results[name] = (pandas_time, numpy_time)

comparison = pd.DataFrame(results, index=['Pandas Time', 'NumPy Time']).T

comparison['Pandas Speed'] = 0
comparison['NumPy Speed'] = 0

for i in range(len(comparison)):
    pandas_time = comparison.iloc[i]['Pandas Time']
    numpy_time = comparison.iloc[i]['NumPy Time']
    
    if pandas_time < numpy_time:
        comparison.at[comparison.index[i], 'Pandas Speed'] = 5
        comparison.at[comparison.index[i], 'NumPy Speed'] = 5 * (pandas_time / numpy_time)
    else:
        comparison.at[comparison.index[i], 'NumPy Speed'] = 5
        comparison.at[comparison.index[i], 'Pandas Speed'] = 5 * (numpy_time / pandas_time)

print(comparison)
```
